```python
import os


pathy = '/users/Seanmullins333/desktop/AAA/Posty'

file_list = os.listdir('/users/Seanmullins333/desktop/AAA/Posty')

for i in file_list[11:]:
    
    final_path = pathy + '/' + i

    y, sr = librosa.load(final_path)

    y = librosa.effects.preemphasis(y)

    S_full, phase = librosa.magphase(librosa.stft(y))
    S_filter = librosa.decompose.nn_filter(S_full,
                                           aggregate=np.median,
                                           metric='cosine',
                                           width=int(librosa.time_to_frames(2, sr=sr)))
    S_filter = np.minimum(S_full, S_filter)
    margin_i, margin_v = 1, 5
    power = 2

    mask_i = librosa.util.softmask(S_filter,
                                   margin_i * (S_full - S_filter),
                                   power=power)

    mask_v = librosa.util.softmask(S_full - S_filter,
                                   margin_v * S_filter,
                                   power=power)

    # Once we have the masks, simply multiply them with the input spectrum
    # to separate the components
    S_foreground = mask_v * S_full
    S_background = mask_i * S_full
    D_foreground = S_foreground * phase
    y_foreground = librosa.istft(D_foreground)
    
    final_file = '/users/Seanmullins333/desktop/AAA/Posty_Vocals' + '/' + i + ' [VOX].wav'
    
    sf.write(final_file, y_foreground, samplerate=22050, subtype='PCM_24')
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

mel_json = {'Artist': [],
             'Mel': [],
              'Label': []}


def extract_mel(folder_path, artist, label):
    for file_name in os.listdir(folder_path):
        try:
            time_point = 31.0  
        
            file_path = os.path.join(folder_path, file_name)

            y, sr = librosa.load(file_path, offset=time_point)

            frame_index = int(time_point * sr)
            win_length = 2048
            hop_length = 512
            
            segment = y[frame_index:frame_index+win_length]
            stft = librosa.stft(segment, n_fft=win_length, hop_length=hop_length)
            magnitude_spectrogram = np.abs(stft)
            spectrogram_db = librosa.amplitude_to_db(magnitude_spectrogram, ref=np.max)

            # Append the feature vector to the list
            mel_json['Mel'].append(spectrogram_db)
            mel_json['Artist'].append(artist)
            mel_json['Label'].append(label)
        except:
            print('error with {}'.format(file_path))   
        
extract_mel('/users/seanmullins333/desktop/AAA/Drake', 'Drake', 0)
extract_mel('/users/seanmullins333/desktop/AAA/Posty', 'Posty', 1)
extract_mel('/users/seanmullins333/desktop/AAA/Weekend', 'Weekend', 2)
```
